# image manip

```python
try:
    logo=np.load("numpy-logo.npy")

    plt.figure(figsize=(8,6))
    plt.subplot(121)
    plt.title("Numpy logo")
    plt.grid(False)
    plt.imshow(logo)

    #dark logo
    dark_logo=1-logo


    plt.figure(figsize=(8,6))
    plt.subplot(122)
    plt.title("Numpy logo dark")
    plt.grid(False)
    plt.imshow(dark_logo)
    
except FileNotFoundError:
    print("logo not found")
```
