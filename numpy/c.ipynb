# Working on data

```python
import numpy as np
import matplotlib.pyplot as plt

# sales data of 5 shops over 4 years 
sales_data=np.array([
    [1,15000,18000,22000,25000],
    [2,12000,14000,16000,19000],
    [3,20000,23000,26000,30000],
    [4,18000,21000,24000,27000],
    [5,16000,18500,20500,23000]
])

print(sales_data.shape,"\n")
print("data of 1st 3 rows:\n",sales_data[:3],"\n")
print("data of all rows without 1st col:\n",sales_data[:,1:],"\n")
```

some queries

```python
# sum of sales per year
print(np.sum(sales_data[:,1:],axis=0),"\n")

#min sales per shop
print(np.min(sales_data[:,1:],axis=1),"\n")

#max sales per year
print(np.max(sales_data[:,1:],axis=0),"\n")

#avg sales per shop
print(np.mean(sales_data[:,1:],axis=1),"\n")

#cumulative per shop
cumsum=np.cumsum(sales_data[:,1:],axis=1)
print(cumsum)
```

# basic plotting

```python
plt.figure(figsize=(8,6))
plt.plot(np.mean(cumsum,axis=0))
plt.title("Mean cumsum over all shops")
plt.xlabel("years")
plt.ylabel("sales")
plt.grid(True)
plt.show()
```

# vectorization and broadcasting
