```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor

iowa_file_path = '../train.csv'

home_data = pd.read_csv(iowa_file_path)
y = home_data.SalePrice

feature_columns = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']
X = home_data[feature_columns]

iowa_model = DecisionTreeRegressor()
iowa_model.fit(X, y)

print("First in-sample predictions:", iowa_model.predict(X.head()))
print("Actual target values for those homes:", y.head().tolist())
```

# Exercises

## Step 1: Split Data
`train_test_split` function to split data.


```python
from sklearn.model_selection import train_test_split

train_X, val_X, train_y, val_y = train_test_split(X,y,random_state=1) 
```

## Step 2: Specify and Fit the Model

Make `DecisionTreeRegressor` model and fit it to the relevant data

```python
iowa_model = DecisionTreeRegressor(random_state=1)
iowa_model.fit(train_X,train_y)
```

## Step 3: Make Predictions with Validation data


```python
val_predictions = iowa_model.predict(val_X)
print(val_predictions)
```

```python
print(val_predictions[:5])
print(val_y.head())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(range(len(val_predictions)),val_predictions,label="pred")
plt.plot(range(len(val_predictions)),val_y,label="actual")
plt.legend()


plt.title("predicted vs actual")
plt.xlabel("index")
plt.ylabel("pred/real data")

plt.grid(True)
plt.show()
```

## Step 4: Calculate the Mean Absolute Error in Validation Data

```python
from sklearn.metrics import mean_absolute_error
val_mae = mean_absolute_error(val_predictions,val_y)

print(val_mae)
```
