# Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[2,3,5,7,11]

plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.title("Simple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

# Diff kinds of plots

```python
categories=['A','B','C']
values=[4,7,2]

plt.plot(x,y)
plt.title("Line Plot")
plt.show()

plt.bar(categories,values)
plt.title("Bar Chart")
plt.show()

plt.scatter(categories,values)
plt.title("Scatter Plot")
plt.show()

data = [1, 2, 2, 3, 3, 3, 4]
plt.hist(data, bins=4,color="Red")
plt.title("Histo")
plt.show()
```

# Customize plots

```python
plt.plot(x, y, color='green', linestyle='--', marker='o')
plt.grid(True)
plt.show()
```

# subplots

```python
fig, axs = plt.subplots(2, 1)  # 2 rows, 1 column
axs[0].plot(x, y)
axs[1].bar(x, y)
plt.show()
```

# Saving plot as img

```python
plt.plot(x, y)
plt.savefig('my_plot.png')  # Saves to file
```

# Practice

```python
fig, axs = plt.subplots(2, 2, figsize=(8, 6))

axs[0, 0].plot([1, 2, 3], [1, 2, 3])
axs[0, 1].bar([1, 2, 3], [3, 2, 1])
axs[1, 0].scatter([1, 2, 3], [3, 1, 2])
axs[1, 1].hist([1, 2, 1, 2, 3, 3, 3])

fig.suptitle("Multiple Subplots")
plt.tight_layout()
plt.show()
```
