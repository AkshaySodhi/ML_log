```python
import pandas as pd

# Create sample data
data = {
    'name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Grace', 'Hannah', 'Ian', 'Jane'],
    'age': [25, 30, 45, 35, 29, 50, 41, 31, 28, 40],
    'experience': [1, 3, 20, 10, 4, 25, 18, 6, 2, 15],
    'salary': [40000, 50000, 120000, 80000, 55000, 150000, 110000, 60000, 42000, 100000],
    'department': ['HR', 'IT', 'Finance', 'IT', 'HR', 'Finance', 'Finance', 'IT', 'HR', 'Finance']
}

df = pd.DataFrame(data)
print(df)
```

## Age histo

```python
import matplotlib.pyplot as plt
import numpy as np

plt.hist(df["age"])
plt.xlabel("Age")
plt.ylabel("Num")
plt.grid(True)
plt.show()
```

## Scatter exp vs salary

```python
plt.scatter(df["experience"],df["salary"])
plt.xlabel("Exp")
plt.ylabel("Salary")
plt.grid(True)
plt.show()
```

## Box plot salary by dept

```python
df.boxplot(column="salary",by="department")
plt.title("sal by dept")
plt.suptitle("")
plt.ylabel("Salary")
plt.show()
```

```python
sorted_df = df.sort_values(by='experience')

plt.plot(sorted_df['experience'], sorted_df['salary'], marker='o')
plt.title('Salary Trend by Experience')
plt.xlabel('Experience (years)')
plt.ylabel('Salary')
plt.grid(True)
plt.show()
```

```python
max_index = df['salary'].idxmax()

plt.scatter(df['experience'], df['salary'], color='purple')
plt.annotate('Highest Salary',
             xy=(df['experience'][max_index], df['salary'][max_index]),
             xytext=(df['experience'][max_index]+1, df['salary'][max_index]+5000),
             arrowprops=dict(arrowstyle='->', color='red'))
plt.title('Salary with Max Annotation')
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.grid(True)
plt.show()
```
